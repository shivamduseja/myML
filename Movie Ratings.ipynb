```python
#FACET GRIDS
```

```python
g=sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
kws=dict(s=50,linewidth=0.5,edgecolor='black')
g.map(plt.scatter,'CriticRating','AudienceRating',**kws)
```

```python
#Controlling axes and adding diagonals
g=sns.FacetGrid(movies,row='Genre',col='Year',hue='Genre')
kws=dict(s=50,linewidth=0.5,edgecolor='black')
g.map(plt.scatter,'CriticRating','AudienceRating',**kws)

for ax in g.axes.flat:
    ax.plot((0,100),(0,100),c='gray',ls='--')
g.add_legend()
```

```python
sns.set_style("white",{"axes.facecolor":"black"})
f,ax=plt.subplots(2,2,figsize=(15,15))

k2=sns.kdeplot(movies.Budget,movies.CriticRating,shade=True,
               shade_lowest=True,ax=ax[0,0],cmap='inferno')
k2b=sns.kdeplot(movies.Budget,movies.CriticRating,ax=ax[0,0],cmap='cool')
k2.set(xlim=(-20,160))


k1=sns.kdeplot(movies.Budget,movies.AudienceRating,shade=True,
               shade_lowest=True,ax=ax[0,1],cmap='inferno')
k1b=sns.kdeplot(movies.Budget,movies.AudienceRating,ax=ax[0,1],cmap='cool')
k1.set(xlim=(-20,160))

z=sns.violinplot(data=movies,x='Year',y='Budget',ax=ax[1,0],
                palette='YlOrRd')

z=sns.boxplot(data=movies,x='Genre',y='CriticRating',ax=ax[1,1],
              palette='Set2',linewidth=3,saturation=0.7)
```

```python
sns.set_style("white")

list1=list()
labels=list()
for i in movies.Genre.unique():
    list1.append((movies[movies.Genre==i].Budget))
    labels.append(i)

fig ,ax=plt.subplots()
fig.set_size_inches(10,5)

h=plt.hist(list1,bins=20,stacked=True,label=labels)
plt.title("Movie Budget Distribution",fontsize=22,color='DarkBlue')
plt.ylabel('Number of Movies',fontsize=15)
plt.xlabel('Budget',fontsize=15)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.legend(fontsize=12,shadow=True)
plt.show()
```
